```python
economic_df.rename(columns={'PER CAPITA INCOME ':'PER CAPITA INCOME', 'PERCENT AGED 16+ UNEMPLOYED':'PERCENT AGED 16plus UNEMPLOYED'
                           ,'PERCENT AGED 25+ WITHOUT HIGH SCHOOL DIPLOMA':'PERCENT AGED 25plus WITHOUT HIGH SCHOOL DIPLOMA'}, inplace =True)
economic_df.columns = economic_df.columns.str.lower().str.replace(' ' , '_')
economic_df.head()
```

### Create new data with select columns

```python
clean_gun_df = gun_df[[ 'Community Area','Date', 'Block',
       'Description', 'Location Description', 'Arrest', 'Domestic', 'Beat',
       'District', 'Ward','Year']]
```

```python
clean_gun_df = clean_gun_df[(clean_gun_df['Year'] >= 2008) & (clean_gun_df['Year'] <= 2012)]
clean_gun_df
```

```python
clean_gun_df.rename(columns={'Community Area':'Community Area Number'}, inplace =True)
clean_gun_df.head()
```

```python
clean_gun_df.columns = clean_gun_df.columns.str.lower().str.replace(' ' , '_')
# clean_gun_df.columns = clean_gun_df.columns.str.lower()
clean_gun_df.head()
```
